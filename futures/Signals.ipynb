## calculate moving averages

```python
df['ema200'] = df.groupby('code')['settle'].apply(pd.ewma, 200, min_periods=200)
df['ema100'] = df.groupby('code')['settle'].apply(pd.ewma, 100, min_periods=100)
df['ema50'] = df.groupby('code')['settle'].apply(pd.ewma, 50, min_periods=50)
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def chart(data, series, x, y, title):

    fig = plt.figure(figsize=(12, 10))

    ax = fig.add_subplot(1, 1, 1)

    ax.spines['right'].set_color('none')
    ax.spines['top'].set_color('none')

    y_min_val = min(data[series[0]])-10
    y_max_val = max(data[series[0]])+10
    ax.set_ylim(y_min_val, y_max_val)

    ax.xaxis.set_ticks_position('bottom')

    plt.title(title)
    plt.xlabel(x)
    plt.ylabel(y)
    
    cmaps = [('Sequential',     ['Blues', 'BuGn', 'BuPu',
                             'GnBu', 'Greens', 'Greys', 'Oranges', 'OrRd',
                             'PuBu', 'PuBuGn', 'PuRd', 'Purples', 'RdPu',
                             'Reds', 'YlGn', 'YlGnBu', 'YlOrBr', 'YlOrRd']),
         ('Sequential (2)', ['afmhot', 'autumn', 'bone', 'cool', 'copper',
                             'gist_heat', 'gray', 'hot', 'pink',
                             'spring', 'summer', 'winter']),
         ('Diverging',      ['BrBG', 'bwr', 'coolwarm', 'PiYG', 'PRGn', 'PuOr',
                             'RdBu', 'RdGy', 'RdYlBu', 'RdYlGn', 'Spectral',
                             'seismic']),
         ('Qualitative',    ['Accent', 'Dark2', 'Paired', 'Pastel1',
                             'Pastel2', 'Set1', 'Set2', 'Set3']),
         ('Miscellaneous',  ['gist_earth', 'terrain', 'ocean', 'gist_stern',
                             'brg', 'CMRmap', 'cubehelix',
                             'gnuplot', 'gnuplot2', 'gist_ncar',
                             'nipy_spectral', 'jet', 'rainbow',
                             'gist_rainbow', 'hsv', 'flag', 'prism'])]
    
    ax.set_color_cycle([plt.cm.Paired(i) for i in np.linspace(0, 1, len(series))])
    
    for s in series:
        # if it's the main series, embolden it
        lw = 2.5 if series.index(s) == 0 else 1.5
        ls = '-' #if series.index(s) == 0 else '--'
        ax.plot(data.index, data[s], linewidth=lw, linestyle=ls)

    return plt 
```

```python
s = ['settle', 'ema50','ema100','ema200']
```

```python
plt = chart(chart_data, s, 'Date', '$', '')
```

###Let's do some model training

```python
from __future__ import print_function
import numpy as np
from scipy import stats
import pandas as pd
import matplotlib.pyplot as plt

import statsmodels.api as sm
from statsmodels.graphics.api import qqplot
```

```python
dta = chart_data[["settle"]].sort()
```

```python
arma_mod20 = sm.tsa.ARMA(dta, (2,0)).fit()
print(arma_mod20.params)
```

```python
arma_mod30 = sm.tsa.ARMA(dta, (3,0))
arma_res30 = arma_mod30.fit()
# .fit()
# arma_mod = sm.tsa.ARMA(y, order=(2,2))
# arma_res = arma_mod.fit(trend='nc', disp=-1)
```

```python
print(arma_res30.summary())
```

```python
fig = plt.figure(figsize=(12,8))
ax = fig.add_subplot(111)
ax = arma_res30.resid.plot(ax=ax);
```

```python
pred_price = arma_mod20.predict('2004', '2015-05-18', dynamic=True)
print(pred_price)
```

```python
fig, ax = plt.subplots(figsize=(12, 8))
# ax = dta.ix['2012':].plot(ax=ax)
fig = arma_mod20.plot_predict('2012', '2015-05-18',ax=ax)
```
